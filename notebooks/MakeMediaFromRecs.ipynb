# Generate Media Specification File From ART Reccomendations File

```python
import pandas as pd
from IPython.display import display
import media_compiler as mc
```

```python
df = pd.read_csv('data/indigoidine/indigoidine_media_Cycle1.csv')
rec_df = pd.read_csv('data/indigoidine/Cycle1_Recommendations.csv',delimiter='\t')
df.tail()
dest_df = df.loc[df['Plate']=='dest_plate'].reset_index()
dest_df = dest_df.loc[:,dest_df.columns != 'index']
dest_df.head()
```

```python
media_cols = [col for col in rec_df.columns if col not in ['Unnamed: 0','Indigoidine']]
rec_df = rec_df.loc[:,media_cols]
```

```python
#Create Compatable Dataframe

for i,row in dest_df.iterrows():
    j = int(i/3)
    dest_df.loc[dest_df.index==i,media_cols] = rec_df.iloc[j].values
    
dest_df.head()
df = pd.concat([df.loc[df['Plate']!='dest_plate'],dest_df])
display(df)
```

```python
df.to_csv('data/indigoidine/indigoidine_media_Cycle2.csv',index=False)
```
